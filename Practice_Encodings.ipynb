```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(
    '/content/drive/MyDrive/train_.csv',
    usecols=[
        # Ordinal columns
        'ExterQual',
        'KitchenQual',
        'BsmtQual',
        'GarageQual',

        # Nominal columns
        'Neighborhood',
        'MSZoning',
        'SaleCondition',

        # Target column
        'SalePrice'
    ]
)
```

```python
df.sample(20)
```

```python
df['BsmtQual'].fillna(df['BsmtQual'].mode()[0], inplace=True)
df['GarageQual'].fillna(df['GarageQual'].mode()[0], inplace=True)
```

```python
df.columns
# MSZoning (Municipal Subdivision Zoning) --> **NOMINAL**
# Type of area where the house is located.
# Examples:
# RL  = Residential Low Density (normal houses)
# RM  = Residential Medium Density (apartments/townhouses)
# FV  = Floating Village Residential
# C   = Commercial area
# This affects house price because some zones are more desirable.

# Neighborhood --> **NOMINAL**
# Specific locality or area name in Ames, Iowa.
# Examples: NAmes, CollgCr, Gilbert, NridgHt.
# Similar to areas like Bandra, Andheri, Powai in Mumbai.
# Location is one of the strongest factors affecting price.

# ExterQual (Exterior Quality)
# Quality of the outside materials of the house.
# Includes walls, bricks, paint, and finishing.
# Values:
# Ex = Excellent
# Gd = Good
# TA = Typical/Average
# Fa = Fair
# Po = Poor

# BsmtQual (Basement Quality)
# Overall quality and height of the basement.
# Basement = room below ground level.
# Uses the same quality scale: Ex > Gd > TA > Fa > Po.
# Missing value (NaN) usually means the house has no basement.

# KitchenQual (Kitchen Quality)
# Quality of the kitchen, including cabinets, countertops, and finish.
# Uses the same quality scale: Ex > Gd > TA > Fa > Po.

# GarageQual (Garage Quality)
# Quality of the garage structure and condition.
# Uses the same quality scale: Ex > Gd > TA > Fa > Po.
# Missing value (NaN) usually means the house has no garage.

# SaleCondition --> **NOMINAL**
# Condition under which the house was sold.
# Examples:
# Normal  = Regular sale
# Abnorml = Abnormal sale (foreclosure, urgent sale, etc.)
# Partial = Newly built house sold before completion
# Sale condition can impact the final price.

# SalePrice --> TARGET
# Final selling price of the house in US dollars.
# This is the target variable that the machine learning model will predict.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('SalePrice', axis=1), df['SalePrice'], test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_cols = ['ExterQual', 'BsmtQual', 'KitchenQual', 'GarageQual']
quality_order = ['Po', 'Fa', 'TA', 'Gd', 'Ex']

# Use the same order for each ordinal column.
oe = OrdinalEncoder(
    categories=[quality_order] * len(ordinal_cols),
    handle_unknown='use_encoded_value',
    unknown_value=-1,
    encoded_missing_value=-1
)
```

```python
# Learn the mapping:
# Po -> 0
# Fa -> 1
# TA -> 2
# Gd -> 3
# Ex -> 4
oe.fit(X_train[ordinal_cols])
```

```python
# Transform the categories into numbers.
X_train[ordinal_cols] = oe.transform(X_train[ordinal_cols])
X_test[ordinal_cols] = oe.transform(X_test[ordinal_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder


nominal_cols = ['Neighborhood', 'MSZoning', 'SaleCondition']

ohe = OneHotEncoder(
    handle_unknown='ignore', # Ignore categories in test data not seen during training
    sparse_output=False # Return a regular NumPy array instead of a sparse matrix
)

# Learn all unique categories from the training data
ohe.fit(X_train[nominal_cols])

# Transform categories into one-hot encoded columns
X_train_nominal = ohe.transform(X_train[nominal_cols])
X_test_nominal = ohe.transform(X_test[nominal_cols])

# Optional: View the names of the generated columns
ohe.get_feature_names_out(nominal_cols)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

ordinal_cols = ['ExterQual', 'BsmtQual', 'KitchenQual', 'GarageQual']
nominal_cols = ['Neighborhood', 'MSZoning', 'SaleCondition']

numeric_cols = [
    col for col in X_train.columns
    if col not in ordinal_cols + nominal_cols
]

X_train_ord = X_train[ordinal_cols].to_numpy()
X_test_ord = X_test[ordinal_cols].to_numpy()

X_train_num = X_train[numeric_cols].to_numpy()
X_test_num = X_test[numeric_cols].to_numpy()

X_train_final = np.concatenate(
    [X_train_ord, X_train_num, X_train_nominal],
    axis=1
)

X_test_final = np.concatenate(
    [X_test_ord, X_test_num, X_test_nominal],
    axis=1
)

model = LinearRegression()
model.fit(X_train_final, y_train)

y_pred = model.predict(X_test_final)

print("R² Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```
