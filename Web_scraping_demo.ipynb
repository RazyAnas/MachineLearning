```python
import pandas as pd
import requests
from bs4 import BeautifulSoup # for webscarping
```

```python
url = 'https://books.toscrape.com/catalogue/page-1.html'
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.google.com/"
}
webpage = requests.get(url=url, headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml') # parse html
```

```python
soup.prettify()
```

```python
for i in soup.find_all('h3'):
  print(i.text.strip()) # proper python string
```

```python
for p in soup.find_all('p', class_='star-rating'):
    rating = p['class'][1]
    print(rating)
```

```python
for i in soup.find_all('p', class_='price_color'):
  print(i.text.strip()[1:])
```

```python
for i in soup.find_all('p', class_='availability'):
  og = i.text.strip()
  text = False
  if og == 'In stock':
    text = True
  print(text)
```

```python
company = soup.find_all('article', class_='product_pod')
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

final = pd.DataFrame()

for j in range(1, 51):   # pages 1 to 50
    url = f"https://books.toscrape.com/catalogue/page-{j}.html"

    headers = {
        "User-Agent": "Mozilla/5.0",
        "Accept-Language": "en-US,en;q=0.9",
        "Referer": "https://www.google.com/"
    }

    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, "html.parser")

    # all book cards on the page
    books = soup.find_all("article", class_="product_pod")

    # temporary lists for one page
    name = []
    rating = []
    price = []
    stock = []

    for book in books:
        # title
        name.append(book.find("h3").find("a")["title"])

        # rating (One, Two, Three, Four, Five)
        rating.append(book.find("p", class_="star-rating")["class"][1])

        # price (remove £ symbol)
        price.append(float(book.find("p", class_="price_color").text[2:]))

        # availability
        availability_text = book.find("p", class_="availability").text.strip()
        stock.append("In stock" in availability_text)

    # DataFrame for this page
    temp_df = pd.DataFrame({
        "name": name,
        "rating": rating,
        "price (£)": price,
        "availability": stock
    })

    # add to final DataFrame
    final = pd.concat([final, temp_df], ignore_index=True)

```
