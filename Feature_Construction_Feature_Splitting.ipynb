# Feature Construction
manual, no math, domain knowledge and intuition required

depends on data we have

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day45-feature-construction-and-feature-splitting/train.csv')[['Age', 'Pclass', 'SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
# in titanic we can make family using SibSp and Parch
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:, 0:4]
y = df.iloc[:, -1]
```

```python
np.mean(cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=20))
```

## Applying Feature Construction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
```

```python
X['isAlone'] = X['Family_size'].apply(
    lambda num: 1 if num == 0 else (1 if num > 1 and num <= 4 else 2)
)
```

```python
X.drop(columns=['SibSp', 'Parch', 'Family_size'], inplace=True)
```

```python
np.mean(cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=20))
```

# Feature Splitting

Data not tidy, more than one data in a single cell

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day45-feature-construction-and-feature-splitting/train.csv')
```

```python
df['Name'].sample(20)
```

```python
df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0] #expand=True means: “Split the text into separate columns.”
```

```python
df['Title']
```

```python
df.groupby('Title')['Survived'].mean().sort_values(ascending=False)
```

```python
X['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0] #expand=True means: “Split the text into separate columns.”
X = pd.get_dummies(X, columns=['Title'], drop_first=True)
```

```python
cross_val_score(
    LogisticRegression(max_iter=1000),
    X,
    y,
    scoring='accuracy',
    cv=20
).mean()
```
