### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

### Step 3. Assign it to a variable called drinks.

```python
variable = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv", sep=",")
```

```python
variable.head(5)
```

### Step 4. Which continent drinks more beer on average?

```python
variable[['continent', 'beer_servings']].groupby('continent').mean().sort_values(by='beer_servings', ascending=False)
```

### Step 5. For each continent print the statistics for wine consumption.

```python
variable[['continent', 'wine_servings']].groupby('continent').describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
variable.groupby('continent').mean(numeric_only=True)
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
variable.groupby('continent').median(numeric_only=True)
```

#### This time output a DataFrame
