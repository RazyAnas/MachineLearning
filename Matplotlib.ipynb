Types of Data
- Numerical Data (Age, weight, Temperature)
- Categorical Data (group, brand, branch, uni, gender)

We will learn 5 types of graphs - 2D plot, scatter plot, bar chart, histogram, pie-chart

- When we plot graph over single column it's called univariant analysis
- When we plot graph over two column it's called biivariant analysis
- When we plot graph over multiple column it's called multivariant analysis



```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2D Line plot

```python
# used for bivariant analyiss
# Categorical -> numerical and numerical -> numerical
# use case - time series data
```

```python
# plotting a simple function

price = [48000, 54000, 57000, 49000, 47000, 45000]
year = [2015, 2016, 2017,2018,2019,2020]

plt.plot(year, price)
```

```python
plt.style.use('default')

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle ='dashed', label='V Kohli', linewidth=2)
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', label='RG Sharma',  linewidth=3)
plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs scored')

plt.legend()
plt.show()


```

```python
plt.style.use('default')

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle ='dashed', label='V Kohli', marker='d', markersize = 10, linewidth=2)
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', label='RG Sharma',  linewidth=3)
plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs scored')

plt.legend(loc='best')
plt.show()


```

# Bar Chart

- Bivariate Analysist
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# x - categories
# y - aggregate operations
```

```python
children = [10, 20, 40, 10, 30]
colors = ['red', 'blue', 'green', 'yellow', 'pink']

plt.bar(colors, children, color = colors)
plt.title('Bacchon ka favourite color')
plt.xlabel('Colors')
plt.ylabel('Children')

plt.grid(visible=True)
```

```python
plt.barh(colors, children, color = colors)
```

```python
df = pd.read_csv("/content/drive/MyDrive/csvs/batsman_season_record.csv")
df.head()
```
