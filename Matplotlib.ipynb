Types of Data
- Numerical Data (Age, weight, Temperature)
- Categorical Data (group, brand, branch, uni, gender)

We will learn 5 types of graphs - 2D plot, scatter plot, bar chart, histogram, pie-chart

- When we plot graph over single column it's called univariant analysis
- When we plot graph over two column it's called biivariant analysis
- When we plot graph over multiple column it's called multivariant analysis



```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2D Line plot

```python
# used for bivariant analyiss
# Categorical -> numerical and numerical -> numerical
# use case - time series data
```

# Bar Chart

- Bivariate Analysist
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# x - categories
# y - aggregate operations
```

```python
children = [10, 20, 40, 10, 30]
colors = ['red', 'blue', 'green', 'yellow', 'pink']

plt.bar(colors, children, color = colors)
plt.title('Bacchon ka favourite color')
plt.xlabel('Colors')
plt.ylabel('Children')

plt.grid(visible=True)
```

```python
plt.barh(colors, children, color = colors)
```

```python
df = pd.read_csv("/content/drive/MyDrive/csvs/batsman_season_record.csv")
df.head()
```

```python
plt.bar(df['batsman'], df['2015'])
# overlapped error
plt.bar(df['batsman'], df['2016'])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
```

```python
# xticks
children = [10, 20, 40, 10, 30]
colors = ['red red red red red', 'blue blue blue blue blue', 'blue', 'yelow yellow', 'pink pink pink pink']

plt.bar(colors, children, color = 'black')
plt.title('Bacchon ka favourite color')
plt.xlabel('Colors')
plt.ylabel('Children')
plt.xticks(rotation='vertical')
plt.grid(visible=True)
```

```python
df
```

```python
# stacked bar chart
plt.bar(df['batsman'], df['2015'], label='2015')
plt.bar(df['batsman'], df['2016'], bottom = df['2015'],  label='2016')
plt.bar(df['batsman'], df['2017'], bottom = df['2015'] + df['2016'],  label='2017')
plt.legend()
```

# Pie Chart

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# Pie chart
data = [23,45,100,20,40]
subjects = ['eng', 'science', 'maths', 'sst', 'hindi']
plt.pie(data, labels=subjects)
plt.show()
```

```python
df = pd.read_csv("/content/drive/MyDrive/csvs/gayle-175.csv")
```

```python
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', colors = ['blue', 'green', 'yellow', 'pink', 'cyan', 'red'])
plt.show()
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode =[0,0,0,0,0,0.1], shadow=True)
plt.show()
```
