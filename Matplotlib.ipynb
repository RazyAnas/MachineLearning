Types of Data
- Numerical Data (Age, weight, Temperature)
- Categorical Data (group, brand, branch, uni, gender)

We will learn 5 types of graphs - 2D plot, scatter plot, bar chart, histogram, pie-chart

- When we plot graph over single column it's called univariant analysis
- When we plot graph over two column it's called biivariant analysis
- When we plot graph over multiple column it's called multivariant analysis

# 3D Surface Plots

```python
x= np.linspace(-10,10,100)
y= np.linspace(-10,10,100)

```

```python
xx, yy= np.meshgrid(x,y)
```

```python
yy.shape
```

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
plot = ax.plot_surface(xx,yy,z,cmap='autumn')

fig.colorbar(plot)
```

```python
z = np.sin(xx) + np.cos(yy)
ax = plt.subplot(projection='3d')
plot = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(plot)
```

```python
z = np.sin(xx) + np.log(yy)
ax = plt.subplot(projection='3d')
plot = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(plot)
```

# Contour plot

3D to 2D

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
# plot = ax.contour(xx,yy,z,cmap='autumn')
plot = ax.contourf(xx,yy,z,cmap='autumn')
fig.colorbar(plot)
```

```python
fig = plt.figure(figsize=(20, 8))

# Left subplot (2D)
ax1 = fig.add_subplot(1, 2, 1)

# Right subplot (3D)
ax2 = fig.add_subplot(1, 2, 2, projection='3d')

z = np.sin(xx) + np.cos(yy)

# 3D surface
surf = ax2.plot_surface(xx, yy, z, cmap='viridis')

# 2D contour
cont = ax1.contourf(xx, yy, z, cmap='viridis')

fig.colorbar(cont, ax=[ax1,ax2])

plt.show()
```
