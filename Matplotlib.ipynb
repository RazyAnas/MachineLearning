Types of Data
- Numerical Data (Age, weight, Temperature)
- Categorical Data (group, brand, branch, uni, gender)

We will learn 5 types of graphs - 2D plot, scatter plot, bar chart, histogram, pie-chart

- When we plot graph over single column it's called univariant analysis
- When we plot graph over two column it's called biivariant analysis
- When we plot graph over multiple column it's called multivariant analysis

```python
batsman = pd.read_csv("/content/drive/MyDrive/csvs/sharma-kohli.csv")
```

```python
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
plt.style.use('default')

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle ='dashed', label='V Kohli', linewidth=2)
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', label='RG Sharma',  linewidth=3)
plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs scored')

plt.legend()
plt.show()


```

```python
plt.style.use('default')

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle ='dashed', label='V Kohli', marker='d', markersize = 10, linewidth=2)
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', label='RG Sharma',  linewidth=3)
plt.title('Rohit Sharma VS Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs scored')

plt.legend(loc='best')
plt.show()


```
