### Challenges in Machine Learning

1. Data Collection

- Easy to get in college project
- Fetch data from an api for big projects - api, web scraping

2. Insufficent Data/Labelled Data

- Algorithm a (100 columns of data) M1, Algorithm b(10,000 columns of data) M2, M2 will perform better.
eg: NLP -> to, too where to use? used different algorithms - realised if we have a lot of data - then algorithms doesnt matter if lot of data.
- insuffiecient data - algorithms matter, we wont get labelled data easily its tedious.

3. Non Representative Data

- Linear regression problem - data gathered isnt complete then what line will draw will differ from the accurate data
- India survey - 2020 t20 worldcup survey - Most will say india will win - its called **sampling noise**
- World survey (large data) - 200 fans (most countries have indians) - most says India will win - its called **Sampling bias**

4. Poor Quality data

- Abrupt, error data - requires lot of work to fix - data quality refining requires lot of time to get better result

5. Irrelevant Features

- Garbage in garbge out
- eg: age | weight | height | location
- location doesnt say much or add any value - remove it
- weight and height **feature engineering** to bmi

6. Overfitting

- ratta maar liya over old data, training data taken too serious
- dont know what when why just remembers old data
- eg: movie ticket price high, we think this place is expensive not only movie ticket

7. Underfitting

- opposite of overfitting
- doesnt think much just make opinion
- neither good for traing data nor for new data

8. Software Integration

- software ke sath integrate krna - lot of os/platforms
- we have to make sure integratable to lot of software platforms
- difficult

9. Offline learning/Deployment

- Offline - train with new data - Update => cumbersome
- Cost in deployment

## Machine Learning Types

Production - server on which our code is going to run


### Batch (offline) Vs Online ML

#### Batch learning

- Convenetional way of training a machine learning model, use all the data at the same time.
- No incremental training in chunks.
- A big data on serever is costly and time taking
- Offline trained
- learnt model is put to production
- data -> model -> test -> server (run)

Problems:
- Not updated data
- Re-train again and again periodically
- Data updated - merge with old data - again trained - put to server in circle in a period
- Depends on amount of data
- Hardware limitation
- Availability

**DAY - 5 - Apr 8 2026**

#### Online Machine Learning
- use our product as much as u can and our product will improve as you use it.
- dynamically trained on new data online.
- done incrementally sequentially after each small chunks.
- Model is in server (online)
- data -> ML algorithm -> test -> server (prediction on new data + Learning) <- inflow of new data
- chatbots - google, alexa, siri
- swift keyboard - type - personalised - online learning
- youtube - particular video clicke - new feed based on that

##### When to use?
- problem nature change frequently -> online learning -> volatile scenarios
- Cost effective -> small chunks -> online
- Faster solution -> online

##### How to implement Online Learning?
- SGDRegressor - online learning - fit and partial fit in sklearn module in python (difficult)
- or dedicated libraries like River - go ot library for doing ML on streamind data
- Vowpal Wabbit - reinforcement and also online learning

##### Learning Rate
- dont make it forget old data forget and also learn new data
- need to set correct learning rate for online learning

#### Out of core learning
- cant use batch learning becoz we cant load a huge data
- in such case use online learning in many batches by converting into smaller data sets -> to model -> test -> server

##### Disadvantage
1. Tricky to use
2. Risky
3. Active monitoring system - anamoly detection - reject bad data
4. Server offline and rollback needed in such case
