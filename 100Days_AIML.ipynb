## Machine Learning Types

Production - server on which our code is going to run


### Batch (offline) Vs Online ML

#### Batch learning

- Convenetional way of training a machine learning model, use all the data at the same time.
- No incremental training in chunks.
- A big data on serever is costly and time taking
- Offline trained
- learnt model is put to production
- data -> model -> test -> server (run)

Problems:
- Not updated data
- Re-train again and again periodically
- Data updated - merge with old data - again trained - put to server in circle in a period
- Depends on amount of data
- Hardware limitation
- Availability

### Online Machine Learning
- use our product as much as u can and our product will improve as you use it.
- dynamically trained on new data online.
- done incrementally sequentially after each small chunks.
- Model is in server (online)
- data -> ML algorithm -> test -> server (prediction on new data + Learning) <- inflow of new data
- chatbots - google, alexa, siri
- swift keyboard - type - personalised - online learning
- youtube - particular video clicke - new feed based on that

##### When to use?
- problem nature change frequently -> online learning -> volatile scenarios
- Cost effective -> small chunks -> online
- Faster solution -> online

##### How to implement Online Learning?
- SGDRegressor - online learning - fit and partial fit in sklearn module in python (difficult)
- or dedicated libraries like River - go ot library for doing ML on streamind data
- Vowpal Wabbit - reinforcement and also online learning

##### Learning Rate
- dont make it forget old data forget and also learn new data
- need to set correct learning rate for online learning

#### Out of core learning
- cant use batch learning becoz we cant load a huge data
- in such case use online learning in many batches by converting into smaller data sets -> to model -> test -> server

##### Disadvantage
1. Tricky to use
2. Risky
3. Active monitoring system - anamoly detection - reject bad data
4. Server offline and rollback needed in such case
