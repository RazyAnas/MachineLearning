```python
# isin()
# Find the most destructive death over batsman in the history of IPL
# basis of Strike Rate = (Number of runs/ Number of balls)/100
# 100 balls me kitne run maare
# criteria: min batsman 200 balls played - over 16 - 20
```

```python
mask = delivery['over'] > 15
death_over = delivery[mask]
all_batsman = death_over.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
x = all_batsman > 200
batsman_list = all_batsman[x].index.tolist()
```

```python
# runs scored by all these each 43 batsman and
# number of balls played by each batsman

mask = death_over['batsman'].isin(batsman_list)
final = death_over[mask]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr.sort_values(ascending=False)
```
