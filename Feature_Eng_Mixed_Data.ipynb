```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day33-handling-mixed-variables/titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with')
```

```python
# Extract numerical part
df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
# pd.to_numeric() is a Pandas function that attempts to convert each value into a numeric type.
# Coerce means “force invalid values to become missing values.”
# Downcasting means: after conversion, Pandas tries to store the numbers in the smallest integer type that can hold them.
```

```python
# Exatract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)
# np.where(condition, value_if_true, value_if_false)
```

```python
df['cabin_num'] = df['Cabin'].str.extract(r'(\d+)') # captures numerical part
df['cabin_cat'] = df['Cabin'].str[0] # captures the first letter
```
