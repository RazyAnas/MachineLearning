# Bias Variance Trade-Off

**Bias**: The inability of a machine learning model to truly capture the relationship in the training data

**Variance**: Difference of fits on various datasets.

Training error nearly equal test error = low variance

**Overfitting**: Training error extremely low, test error extremely high

**Underfitting**: Training error High. Not able to find relation in training data.

***BEST: low bias and low variance***

Ways:

1. Regularising
2. Bagging
3. Boosting
